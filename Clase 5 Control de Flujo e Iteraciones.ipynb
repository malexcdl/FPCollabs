# CICLO WHILE

```python
vc = 0
while vc <= 7:
  vc += 1 # vc = vc + 1
  print("La variable de control tiene un valor de:", vc)
```

```python
vc = 0
while vc <= 7:
  vc += 1 # vc = vc + 1
  print("La variable de control tiene un valor de:", vc)
else:
  print("Ha terminado el ciclo while, y el valor de la variable de control terminó en:", vc)
```

```python
# Instrucción para romper el ciclo break:
contador = 0
while contador <= 10:
  contador += 1
  if contador == 6:
    print("Se interrumpe el bucle cuando el contador vale:", contador)
    break
  print("contador vale", contador)
else:
  print("Ha terminado el ciclo while")
```

```python
# Instrucción para saltarse una iteración y seguir continue.
contador = 0
while contador <= 10:
  contador += 1
  if contador == 5 or contador == 8:
    print("Se continua con la siguiente iteración")
    continue
  print("contador vale", contador)
else:
  print("Ha terminado el ciclo while")
```

```python
print("PROGRAMA PARA REALIZAR OPERACIONES MATEMÁTICAS BÁSICAS\n")
while True:
  print("""¿Qué operación desea realizar? Escribe el número de la opción deseada:
    1.) Sumar dos números
    2.) Restar dos números
    3.) Multiplicar dos números
    4.) Dividir dos números
    5.) Salir del Programa""")
  opcion = input()
  if opcion == "1":
    n1 = float(input("Ingresa el valor del primer número: "))
    n2 = float(input("Ingresa el valor del segundo número: "))
    print("El resultado de la suma de los dos números es: ", n1 + n2)
  elif opcion == "2":
    n1 = float(input("Ingresa el valor del primer número: "))
    n2 = float(input("Ingresa el valor del segundo número: "))
    print("El resultado de la resta de los dos números es: ", n1 - n2)
  elif opcion == "3":
    n1 = float(input("Ingresa el valor del primer número: "))
    n2 = float(input("Ingresa el valor del segundo número: "))
    print("El resultado de la multiplicación de los dos números es: ", n1 * n2)
  elif opcion == "4":
    n1 = float(input("Ingresa el valor del primer número: "))
    n2 = float(input("Ingresa el valor del segundo número: "))
    if n2 == 0:
      print("Error, no se puede dividir por cero.")
    else:
      print("El resultado de la división de los dos números es: ", n1 / n2)
  elif opcion == "5":
    print("Saliendo del programa... Gracias por utilizar nuestro servicio")
    break
  else:
    print("La opción seleccionada no es válida, intenta de nuevo")
  print("")

```
