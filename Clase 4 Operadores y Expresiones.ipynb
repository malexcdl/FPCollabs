# Booleanos

```python
verdadero = True
falso = False

# OR
print("Con el operador booleano or, tenemos que: ")
print("verdadero or verdadero: ", verdadero or verdadero)
print("verdadero or falso: ", verdadero or falso)
print("falso or verdadero: ", falso or verdadero)
print("falso or falso: ", falso or falso)
```

```python
# AND
print("Con el operador booleano and, tenemos que: ")
print("verdadero and verdadero: ", verdadero and verdadero)
print("verdadero and falso: ", verdadero and falso)
print("falso and verdadero: ", falso and verdadero)
print("falso and falso: ", falso and falso)
```

```python
# NOT
print("Con el operador booleano not, tenemos que: ")
print("not verdadero es:", not verdadero)
print("not false es:", not falso)
```

# Bitwise

```python
a = 2
b = 4
print("El valor binario de a es:", bin(a)[2:].zfill(4))
print("El valor binario de b es:", bin(b)[2:].zfill(4))
```

```python
# and
print("And a & b a nivel de bits es:", a & b)
```

```python
# or
print("Or a | b a nivel de bits es:", a | b)
```

```python
# not
print("Not a y not b a nivel de bits es:", ~a, ~b)
print("-" + bin(~a)[3:].zfill(4))
print("-" + bin(~b)[3:].zfill(4))
```

```python
# xor
print("Xor ^ a nivel de bits es:", a ^ b)
print(bin(a ^ b)[2:].zfill(4))
```

```python
# right shift
print(bin(a)[2:].zfill(4))
print(bin(a >> 1)[2:].zfill(4))
print("Right Shift a nivel de bits desplazando 1 posición a la derecha es:", a >> 1 )
```

```python
# left shift
print(bin(a)[2:].zfill(4))
print(bin(a << 2)[2:].zfill(4))
print("Left Shift a nivel de bits desplazando 2 posiciones a la izquierda es:", a << 2 )
```
