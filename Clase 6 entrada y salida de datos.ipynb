```python
# Programa para calcular el área de un cuadrado.
print("Este programa calculará el área de un Cuadrado")
lado  = float(input("Ingrese el valor de lado del cuadrado: "))
print("""Ahora seleccione la unidad de medida:
1.) mm
2.) cm
3.) m
4.) Km""")
area = lado * lado
unidad_distancia = int(input(""))
if unidad_distancia == 1:
  print("El área del cuadrado es:", area, "milímetros cuadrados mm^2")
elif unidad_distancia == 2:
  print("El área del cuadrado es:", area, "centímetros cuadrados cm^2")
elif unidad_distancia == 3:
  print("El área del cuadrado es:", area, "metros cuadrados m^2")
elif unidad_distancia == 4:
  print("El área del cuadrado es:", area, "Kilómetros cuadrados Km^2")
else:
  print("La unidad de longitud seleccionada no es válida")
```

```python
# Programa para llenar una lista con n valores.
print("Este programa le permite crear una lista con la cantidad de valores deseados")
lista = []
elementos = int(input("Ingrese la cantidad de elementos que desea tener en la lista: "))
for i in range (elementos):
  lista.append(input(f"introduce el valor {i + 1}: "))
print("La lista resultante es:")
print(lista)
```

```python
# Este programa permite ingresar dos números y devolver varios resultados con diferentes operadores.
print("Este programa permite utilizar diferentes operadores en Python a partir de 2 números solicitados Ingresados:")
num1 = int(input("Ingrese el valor del primer número: "))
num2 = int(input("Ingrese el valor del segundo número: "))
print("")
print("La suma de los números es:", num1 + num2)
print("La resta de los dos números es:", num1 - num2)
print("La multiplicación de los dos números es:", num1 * num2)
print("La división de los dos números es:", num1 / num2)
print("La división entera de los dos números es:", num1 // num2)
print("El módulo de los dos números es:", num1 % num2)
if num1 == num2:
  print("Los números son iguales.")
elif num1 > num2: 
  print("El primer número es mayor que el segundo")
elif num1 < num2: 
  print("El primer número es menor que el segundo")
if num1 != num2:
  print("Los números son diferentes")
print("El número 1 en binario es:", bin(num1))
print("El número 2 en binario es:", bin(num2))
print("El número 1 en hexadecimal es:", hex(num1))
print("El número 2 en hexadecimal es:", hex(num2))
```
