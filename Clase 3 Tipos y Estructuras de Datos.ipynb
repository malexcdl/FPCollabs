#LISTAS

```python
# Las listas en python, son arreglos de una dimensión que permiten almacenar distintos tipos de datos, para definir las listas se utilizan los corchetes y sus elementos son separados por coma (,)

palabras = ["Hola,", "esta", "oración", "está", "almacenada", "en", "una", "lista"]
booleanos = [True, False]
enteros = [-9, 8, -7, 6, -5, 4]
flotantes = [1.2, 98.1, -2.54, 4.56, -20.0]
mixta = ["Hola mundo", True, 18, -5.1, [1, 2]]
print(mixta[-1])
print(flotantes[-2])
print(palabras[-4])
print(palabras[3][3])
# print(palabras[4][0:7])
bodega = palabras[4][0:7]
print(bodega)
print(mixta[-1][-2])
mixta[0] = "Hello World"
print(mixta)
mixta[-2] = "9"
print(mixta)
print(type(mixta[-2]))
mixta[-2] = 9
print(mixta)
print(type(mixta[-2]))
print(type(enteros))
```
